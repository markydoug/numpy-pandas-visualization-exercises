```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
x = list(range(1, 4))
y = [0*n +1 for n in x]
z = [np.log(n) for n in x]
a = [n for n in x]
b = [n*np.log(n) for n in x]
c = [n ** 2 for n in x]
d = [2 ** n for n in x]
e = [math.factorial(n) for n in x]
f = [n ** n for n in x]

fig, ax = plt.subplots()

plt.plot(x, y, c='blue', label='$O(1)$')
plt.plot(x, z, c='orange', label='$O(log n)$')
plt.plot(x, a,  c='green', label='$O(n)$')
plt.plot(x, b, c='red', label='$O(n log n)$')
plt.plot(x, c, c='yellow', label='$O(n^2)$')
plt.plot(x, d, c='purple', label='$O(2^n)$')
plt.plot(x, e,  c='green', label='$O(n!)$')
plt.plot(x, f, c='indigo', label='$O(n^n)$')

plt.legend(loc='upper right')
plt.title("Big O Notation")
plt.xlabel('Elements')
plt.ylabel('Operations')

plt.show()
```
