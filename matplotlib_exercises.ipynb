### Use matplotlib to plot the following equation:

#### y = x^2-x+2

```python
x = list(range(-25, 25))
y = [n ** 2 - n + 2 for n in x]
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('$y = x^2 - x + 2$')
plt.annotate('Origin', xy = (0, 0), xytext=(5, 400), arrowprops = dict(facecolor = 'gray')  )
plt.show()
```

### Create and label 4 separate charts for the following equations (choose a range for x that makes sense):

#### y = âˆšx

```python
x = list(range(0, 49))
y = [np.sqrt(n) for n in x]
fig, ax = plt.subplots()
ax.set_title('$\sqrt{x}$')
ax.plot(x, y)
plt.show()
```

#### y = x^3

```python
x = list(range(0, 51))
y = [n ** 3 for n in x]
fig, ax = plt.subplots()
ax.set_title('$x^3$')
ax.plot(x, y)
plt.show()
```

#### y = 2^x

```python
x = list(range(0, 30))
y = [2 ** n for n in x]
fig, ax = plt.subplots()
ax.set_title('$2^x$')
ax.plot(x, y)
plt.ylim(-10000,100000)
plt.show()
```

#### y = 1/(x+1)

```python
x = list(range(0, 51))
y = [(1/(n+1)) for n in x]
fig, ax = plt.subplots()
ax.set_title('$1/(x+1)$')
ax.plot(x, y)
plt.show()
```

### Combine the figures you created in the last step into one large figure with 4 subplots.

```python
x = list(range(0, 51))
y = [np.sqrt(n) for n in x]
z = [n ** 3 for n in x]
a = [2 ** n for n in x]
b = [(1/(n+1)) for n in x]

fig, ax = plt.subplots(2,2, sharex=True)

ax[0,0].plot(x, y)
ax[0,0].set_title('$\sqrt{x}$')
ax[1,0].plot(x, z)
ax[1,0].set_title('$x^3$')
ax[1,1].plot(x, a)
ax[1,1].set_title('$2^x$')
ax[0,1].plot(x, b)
ax[0,1].set_title('$1/(x+1)$')


fig.suptitle('Yay Graphs!')

plt.style.use('seaborn')
plt.show()
```

### Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

```python
x = list(range(0, 50))
y = [np.sqrt(n) for n in x]
z = [n ** 3 for n in x]
a = [2 ** n for n in x]
b = [(1/(n+1)) for n in x]

fig, ax = plt.subplots()

ax.plot(x, y, c='blue', label='$\sqrt{x}$')
ax.plot(x, z, c='orange', label='$x^3$')
ax.plot(x, a,  c='green', label='$2^x$')
ax.plot(x, b, c='red', label='$1/(x+1)$')

ax.legend(loc='upper right', frameon=True)
ax.set_title("Fun Graph Practice")
ax.set_xlabel('$x$')


ax.set_ylim(0,11)
ax.set_xlim(0,11)
plt.show()
```
