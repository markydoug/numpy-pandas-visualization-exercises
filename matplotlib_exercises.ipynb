### Use matplotlib to plot the following equation:

#### y = x^2-x+2

```python
x = list(range(-23, 25))
y = [n ** 2 - n + 2 for n in x]
fig, ax = plt.subplots()
ax.plot(x, y)
plt.title('$y = x^2 - x + 2$')
plt.xlabel('$x$')
plt.ylabel('$x^2$ - x + 2$')
plt.show()
```

### Combine the figures you created in the last step into one large figure with 4 subplots.

```python
x = list(range(0, 10))
y = [np.sqrt(n) for n in x]
z = [n ** 3 for n in x]
a = [n * 2 for n in x]
b = [(1/(n+1)) for n in x]
fig, ax = plt.subplots(4)
ax[0].plot(x, y)
ax[1].plot(x, z)
ax[2].plot(x, a)
ax[3].plot(x, b)
plt.show()
```

### Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

```python
x = list(range(0, 3))
y = [np.sqrt(n) for n in x]
z = [n ** 3 for n in x]
a = [n * 2 for n in x]
b = [(1/(n+1)) for n in x]

fig, ax = plt.subplots()

plt.plot(x, y, c='blue', label='$âˆšx$')
plt.plot(x, z, c='orange', label='$n^3$')
plt.plot(x, a,  c='green', label='$x$')
plt.plot(x, b, c='red', label='$1/(x+1)$')

plt.legend(loc='upper right')
plt.title("Fun Graph Practice")
plt.xlabel('$x$')

plt.show()
```
